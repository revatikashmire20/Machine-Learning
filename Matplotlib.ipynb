```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#scatter plot
```

```python
x = np.arange(0,20)
y = np.arange(30,50)
```

```python
plt.scatter(x,y,c="g",marker = '*')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title('Scatter plot using Matplotlib')

plt.show()
```

```python
plt.plot(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title('Line Graph using Matplotlib')
plt.show()
```

```python
#pie chart
```

```python
x = [900,250,400,60]
labels = ['China','US','Russia','India']
colors = ['r','g','b','y']
```

```python
plt.pie(x,labels=labels, colors=colors, shadow=True)
plt.title("This is as PIE Chart")
plt.show()
```

```python
e=(0.1,0,0,0)
plt.pie(x, explode = e, labels=labels, colors=colors, shadow=True)
plt.title("This is as PIE Chart")
plt.show()
```

```python
# sine wave from 0 to 2 pie
import numpy as np
x1 = np.arange(0,2*np.pi,0.1)
y1 = np.sin(x1)

plt.plot(x1,y1,'r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Sine wave')
plt.show()
```

```python
x = np.array([24,85,4,44,52,7,75,63,11,18,50,78,30,35])
plt.hist(x)
plt.xlabel('X values')
plt.ylabel('No of values')
plt.title('historgram')
plt.show()
```

```python
x = np.array([24,85,4,44,52,7,75,63,11,18,50,78,30,35])
plt.hist(x,bins = 10)
plt.xlabel('X values')
plt.ylabel('No of values')
plt.xticks(np.arange(0,90,5))
plt.title('historgram')
plt.show()
```

```python
x = np.arange(10)
y = x**2

# to change the fig size
plt.figure(figsize = (10,10))

# first plot
plt.subplot(2,2,1)
plt.plot(x,y,'g--')
plt.title('fig 1')

#second plot
plt.subplot(2,2,2)
plt.plot(x,y,'r*--')
plt.title('fig 2')

#third plot
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.title('fig 3')

#fourth subplot
plt.subplot(2,2,4)
plt.plot(x,y,'go')
plt.title('fig 4')


plt.show()
```
