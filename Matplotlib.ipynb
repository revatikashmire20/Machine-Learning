```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#scatter plot
```

```python
x = np.arange(0,20)
y = np.arange(30,50)
```

```python
plt.scatter(x,y,c="g",marker = '*')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title('Scatter plot using Matplotlib')

plt.show()
```

```python
# line plot
```

```python
plt.plot(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title('Line Graph using Matplotlib')
plt.show()
```

```python
df = [20,15,25,10,100]
```

```python
plt.boxplot(df)
plt.show()
```

```python
x = ['5th','6th','7th']
#strength
y = [40,50,47]
```

```python
plt.bar(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title('Bar Graph using Matplotlib')
plt.show()
```

```python
#pie chart
```

```python
x = [900,250,400,60]
labels = ['China','US','Russia','India']
colors = ['r','g','b','y']
```

```python
plt.pie(x,labels=labels, colors=colors, shadow=True)
plt.title("This is as PIE Chart")
plt.show()
```

```python
e=(0.1,0,0,0)
plt.pie(x, explode = e, labels=labels, colors=colors, shadow=True)
plt.title("This is as PIE Chart")
plt.show()
```
